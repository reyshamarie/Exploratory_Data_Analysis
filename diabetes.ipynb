# Skew

```python
skew = diabetes.skew()
print(skew)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(skew.index, skew.values, marker='o', linestyle='-')
plt.xlabel('Attributes')
plt.ylabel('Skewness')
plt.title('Skewness of Attributes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# preg

Positively Skewed: A right-skew in pregnancies could suggest that most women in the dataset have had a few pregnancies. Higher numbers of pregnancies are often associated with an increased risk of gestational diabetes, which can contribute to overall diabetes risk.

# plas

Slightly Positively Skewed: Glucose levels are a critical factor in diabetes. A slight right skew indicates that while most individuals have glucose levels within a normal range, there are some individuals with higher glucose levels, which is a key indicator of diabetes.

# pres

Negatively Skewed: This indicates that most individuals in the dataset have higher blood pressure, with fewer individuals having lower blood pressure. High blood pressure is often associated with diabetes and can lead to complications in diabetic patients.

# skin

Near Zero Skewness: Skin thickness measurements are relatively symmetrically distributed, meaning there is no extreme tail.

# test

Strongly Positively Skewed: A heavy right tail indicates that while most individuals have normal or lower insulin levels, a smaller number have significantly elevated levels. Elevated insulin levels are a strong indicator of insulin resistance, a hallmark of Type 2 diabetes.

# mass

Slightly Negatively Skewed: High BMI is a well-known risk factor for diabetes, particularly Type 2 diabetes, as it is often associated with obesity.

# pedi

Positively Skewed: This attribute measures the likelihood of diabetes based on family history, so a positive skew suggests that there are a few individuals with a significant family history of diabetes.

# Age

Positively Skewed: Age is positively skewed, As age increases, the risk of developing Type 2 diabetes also increases, so this skewness may indicate that the dataset contains more younger individuals who are at lower risk for diabetes.

# class

Slightly Positively Skewed: The slight positive skew in the outcome (likely representing the presence of diabetes) indicates that most individuals in the dataset do not have diabetes, but there is a small group with a higher prevalence of the condition.
